```python
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

```python
df['EthnicGroup'] = df['EthnicGroup'].fillna('Not Available')
df.head()
```

```python
df['ParentEduc'] = df['ParentEduc'].fillna('Not Available')
df['TestPrep'] = df['TestPrep'].fillna('Not Available')
df['ParentMaritalStatus'] = df['ParentMaritalStatus'].fillna('Not Available')
df['PracticeSport'] = df['PracticeSport'].fillna('Not Availabl')
df['IsFirstChild'] = df['IsFirstChild'].fillna('Not Available')
df['TransportMeans'] = df['TransportMeans'].fillna('Not Available')
df.head()
```

```python
df['NrSiblings'] = df['NrSiblings'].fillna(0)
df['WklyStudyHours'] = df['WklyStudyHours'].fillna(0)
df.head()
```

```python
# First, look at the unique values to confirm the patterns
print(df['WklyStudyHours'].unique())
```

```python
# Map string categories to representative numeric values
study_hour_mapping = {
    '< 5': 3,
    '5 - 10': 7,
    '> 10': 12
}

# Replace string with numeric values
df['WklyStudyHours'] = df['WklyStudyHours'].map(study_hour_mapping)
```

```python
# Replace NaN or infinite values with a default value (e.g., 0 or mean of the column)
df['WklyStudyHours'] = df['WklyStudyHours'].replace([np.inf, -np.inf], np.nan)

# Replace NaN values with 0 or any other value that makes sense for your context
df['WklyStudyHours'] = df['WklyStudyHours'].fillna(0)  # Replacing NaN with 0, or choose another strategy

# Now convert to integer type
df['WklyStudyHours'] = df['WklyStudyHours'].astype(int)

# Create StudyLevel column with encoded values
def categorize_study_hours(hours):
    if hours < 5:
        return 0  # Low
    elif 5 <= hours <= 10:
        return 1  # Medium
    else:
        return 2  # High

df['StudyLevel'] = df['WklyStudyHours'].apply(categorize_study_hours)

# Optional: label them for better readability
df['StudyLabel'] = df['StudyLevel'].map({0: 'Low', 1: 'Medium', 2: 'High'})

# Preview the result (fixing the column name 'StudyLevelLabel' to 'StudyLabel')
print(df[['WklyStudyHours', 'StudyLevel', 'StudyLabel']].head(10))
```

```python
# Pie chart for StudyLabel
study_label_counts = df['StudyLabel'].value_counts()

# Plotting
plt.figure(figsize=(7, 7))
study_label_counts.plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'orange'], startangle=90, legend=False)

# Add title
plt.title('Study Level Distribution')

# Show the plot
plt.show()
```

```python
master_degree_1 = df[df['ParentEduc'] == "master's degree"]
some_high_school_2 = df[df['ParentEduc'] == "some high school"]

# Calculate average scores for each subject and education level
avg_scores = {
    'MathScore': [master_degree_1['MathScore'].mean(), some_high_school_2['MathScore'].mean()],
    'ReadingScore': [master_degree_1['ReadingScore'].mean(), some_high_school_2['ReadingScore'].mean()],
    'WritingScore': [master_degree_1['WritingScore'].mean(), some_high_school_2['WritingScore'].mean()]
}

# Calculate and print percentage margins for each subject
for subject, scores in avg_scores.items():
    margin_percentage = ((scores[0] - scores[1]) / scores[1]) * 100
    print(f"The percentage margin of {subject} between Master's degree and Some High School is: {margin_percentage:.2f}%")
```

```python
# Visualization using a bar chart
subjects = list(avg_scores.keys())
margins = [((scores[0] - scores[1]) / scores[1]) * 100 for scores in avg_scores.values()]

plt.figure(figsize=(6,6))
plt.bar(subjects, margins, color=['skyblue', 'orange', 'red'])
plt.title('Percentage Margin between Master\'s Degree and Some High School')
plt.xlabel('Subject')
plt.ylabel('Percentage Margin (%)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding value labels to the bars
for i, v in enumerate(margins):
    plt.text(i, v, f'{v:.2f}%', ha='center', fontweight='bold')

plt.show()
```

```python
# Step 1: Create TotalScore just to define 'Pass'
df["TotalScore"] = df["MathScore"] + df["ReadingScore"] + df["WritingScore"]

# Step 2: Create Pass/Fail column based on TotalScore
df["Pass"] = df["TotalScore"].apply(lambda x: 1 if x >= 150 else 0)

# Step 3: Drop TotalScore since we're not using it anymore
df = df.drop(columns=["TotalScore"])

# Step 4: Drop any derived study level/label columns (optional cleanup)
df = df.drop(columns=[col for col in df.columns if 'StudyLabel' in col or 'StudyLevel' in col])

# Step 5: Convert categorical features to dummy variables
df = pd.get_dummies(df, drop_first=True)
```

```python
# Separate the data into two classes
fail_df = df[df['Pass'] == 0]  # students who failed
pass_df = df[df['Pass'] == 1]  # students who passed

# Randomly sample 600 from each class
fail_sample = fail_df.sample(n=600, random_state=42)
pass_sample = pass_df.sample(n=600, random_state=42)

# Combine to form a balanced dataset
balanced_df = pd.concat([fail_sample, pass_sample]).sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# Import necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Step 1: Split features and target
X_bal = balanced_df.drop(["Pass"], axis=1)
y_bal = balanced_df["Pass"]

# Step 2: Train-test split
X_train_bal, X_test_bal, y_train_bal, y_test_bal = train_test_split(X_bal, y_bal, test_size=0.2, random_state=42)

# Step 3: Train Random Forest classifier (no SMOTE)
clf = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42)
clf.fit(X_train_bal, y_train_bal)

# Step 4: Predict and evaluate
y_pred_bal = clf.predict(X_test_bal)

print("Accuracy:", accuracy_score(y_test_bal, y_pred_bal))
print(classification_report(y_test_bal, y_pred_bal))
print("Confusion Matrix:\n", confusion_matrix(y_test_bal, y_pred_bal))

# Step 5: Visualize the Confusion Matrix
cm = confusion_matrix(y_test_bal, y_pred_bal)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

# Step 1: Prepare features and target from the FULL dataset
# Do NOT drop MathScore, ReadingScore, or WritingScore since they were used during training
X_full = df.drop(["Pass"], axis=1)
y_full = df["Pass"]

# Step 2: Predict using the trained model
y_pred_full = clf.predict(X_full)

# Step 3: Evaluate
print("Accuracy on full dataset:", accuracy_score(y_full, y_pred_full))
print(classification_report(y_full, y_pred_full))
print("Confusion Matrix:\n", confusion_matrix(y_full, y_pred_full))

# Step 4: Visualize the Confusion Matrix
cm = confusion_matrix(y_full, y_pred_full)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```
