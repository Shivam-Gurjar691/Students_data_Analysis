```python
# Create Total Score
df["TotalScore"] = df["MathScore"] + df["ReadingScore"] + df["WritingScore"]

# Create Pass/Fail column
df["Pass"] = df["TotalScore"].apply(lambda x: 1 if x >= 150 else 0)

# Convert categorical columns to dummy variables
df = pd.get_dummies(df, drop_first=True)

```

```python
# Improved Random Forest Regressor with tuned parameters
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Feature-target split
X_reg = df.drop(["MathScore", "ReadingScore", "WritingScore", "TotalScore", "Pass"], axis=1)
y_reg = df["TotalScore"]

# Train-test split
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Initialize and train the model
rf_regressor = RandomForestRegressor(n_estimators=300, max_depth=10, random_state=42)
rf_regressor.fit(X_train_reg, y_train_reg)

# Predictions
y_pred_reg = rf_regressor.predict(X_test_reg)

# Evaluation
print("\n--- Improved Random Forest Regression ---")
print("MAE:", mean_absolute_error(y_test_reg, y_pred_reg))
print("MSE:", mean_squared_error(y_test_reg, y_pred_reg))
print("R2 Score:", r2_score(y_test_reg, y_pred_reg))
```

```python
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Features and target
X_clf = df.drop(["MathScore", "ReadingScore", "WritingScore", "TotalScore", "Pass"], axis=1)
y_clf = df["Pass"]

# Train-test split
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)

# --- SMOTE for balancing classes ---
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train_clf, y_train_clf)

# Train classifier on balanced data
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_resampled, y_resampled)

# ---------- Default prediction (threshold = 0.5) ----------
y_pred_clf = rf_classifier.predict(X_test_clf)

# Evaluation at default threshold
print("\n--- Improved Random Forest Classification with SMOTE ---")
print("Accuracy:", accuracy_score(y_test_clf, y_pred_clf))
print(classification_report(y_test_clf, y_pred_clf))

# ---------- Threshold tuning ----------
# Get predicted probabilities for class 1
y_proba = rf_classifier.predict_proba(X_test_clf)[:, 1]

# Set a custom threshold (e.g., 0.3)
threshold = 0.4
y_pred_thresh = (y_proba >= threshold).astype(int)

# Evaluate at custom threshold
print(f"\n--- Classification Report @ Threshold = {threshold} ---")
print(classification_report(y_test_clf, y_pred_thresh))
print("Confusion Matrix:\n", confusion_matrix(y_test_clf, y_pred_thresh))
print("Accuracy:", accuracy_score(y_test_clf, y_pred_thresh))
```
